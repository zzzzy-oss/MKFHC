```python
import json
from docx import Document

def json_to_word(json_file_path, output_word_path):
    # 创建一个新的Word文档
    doc = Document()

    # 打开并加载JSON文件
    with open(json_file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 遍历JSON中的每个条目，并为每个条目（文章）添加标题
    for index, item in enumerate(data[:6], start=1):
        if isinstance(item, list) and len(item) > 0:
            # 添加文章标题
            doc.add_heading(f'Article {index}', level=1)
            
            # 获取列表中的第一个元素，即文章内容
            paragraph_content = item[0].strip()  # 去除首尾空白字符
            
            # 将文章内容分割成段落，这里我们使用两个换行符'\n\n'作为分隔符
            paragraphs = paragraph_content.split('\n\n')
            for p in paragraphs:
                if p:  # 如果段落不是空的
                    # 添加段落到Word文档中
                    doc.add_paragraph(p.strip())
                    # 在段落之间添加一个空行
                    doc.add_paragraph('')

    # 保存Word文档
    doc.save(output_word_path)
    print(f"文档已成功保存至: {output_word_path}")

# 使用示例
json_file_path = '/home/fzus/zy/experiment/data/webqa_start0_endNone_base_ret10/fusionLma70B.json'  # 替换为你的JSON文件路径
output_word_path = '/home/fzus/zy/experiment/data/webqa_start0_endNone_base_ret10/fopLma70B.docx'  # 替换为你想要保存的Word文档路径
json_to_word(json_file_path, output_word_path)
```
